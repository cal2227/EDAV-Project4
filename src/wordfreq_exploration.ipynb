```python
from matplotlib import pyplot as plt

import numpy as np

import nltk

import utils
```

```python
%matplotlib inline
```

```python
suas = utils.get_suas_1970()
```

```python
cfd1_words = ['democrat', 'republican', 'politics']
cfd1 = utils.gen_ConditionalFreqDist(suas, cfd1_words)
```

```python
cfd1.plot()
```

```python
partisan_words = ['democrat', 'republican', 'politic', 'party', 'agenda', 'debate', 'disagree', 'partisan', 
                  'gridlock', 'divide', 'veto']
cfd2 = utils.gen_ConditionalFreqDist(suas, partisan_words)
```

```python
cfd2.plot()
```

```python
cfd2_agg_counts = []
cfd2_agg_years = []
for sua in suas:
    cfd2_agg_years.append(sua['year'])
    cfd2_agg_counts.append(0)
    for w in nltk.word_tokenize(sua['text']):
        contains_any = np.any([w.lower().startswith(target) \
            for target in partisan_words])
        if contains_any:
            cfd2_agg_counts[-1] += 1
```

```python
plt.scatter(cfd2_agg_years, cfd2_agg_counts)
```
