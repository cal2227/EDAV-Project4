```python
from matplotlib import pyplot as plt

import nltk

import utils
```

```python
%matplotlib inline
```

```python
suas = [sua for sua in utils.get_suas() if int(sua['year']) >= 1970]
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, sua['year'])    
    for sua in suas    
    for w in nltk.word_tokenize(sua['text'])
    for target in ['america', 'citizen', 'people']
    if w.lower().startswith(target))
```
