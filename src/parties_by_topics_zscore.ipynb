```python
import pandas as pd
import seaborn as sb
import numpy as np
import nltk
import utils
```

```python
%matplotlib inline
```

```python
stemmed_keywords = utils.get_stemmed_keywords()
```

```python
df = pd.read_json(utils.suas_with_topic_frequencies)
```

```python
# drop jimmy carter written sua
df = df[df['year'] != 1981]
```

```python
df['environment'].mean(), df['environment'].std()
```

```python
for topic in stemmed_keywords.keys():
    df[topic] = (df[topic] -df[topic].mean()) / df[topic].std()
```

```python
for topic in stemmed_keywords.keys():
    sb.lmplot('year', topic, data=df, hue='political_party', fit_reg=True)
```
