```python
from matplotlib import pyplot as plt

import nltk

import utils
```

```python
%matplotlib inline
```

```python
suas = utils.get_suas_1970()
```

```python
cfd = nltk.ConditionalFreqDist(
    (target, sua['year'])    
    for sua in suas    
    for w in nltk.word_tokenize(sua['text'])
    for target in ['immigra', 'border']
    if w.lower().startswith(target))
```

```python
cfd.plot()
```

```python
cfd2 = nltk.ConditionalFreqDist(
    (target, sua['year'])    
    for sua in suas    
    for w in nltk.word_tokenize(sua['text'])
    for target in ['mexico','china','india']
    if w.lower().startswith(target))
```

```python
cfd2.plot()
```
